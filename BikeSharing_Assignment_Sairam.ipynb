# Transforming the Data


```python
# The numerical variables will be dropped  since instant is the just the serial number of the record
# dteday is redundant because  the required data for analysis is contained in mnth,yr 
# casual + registered = cnt

Bike_df = Bike_df.drop(columns=['instant','dteday','casual','registered'])
```

```python
# As per data dictionary, following are the categorical variables has following values.
# 1. yr has two values 0 for 2018 and 1 for 2019.
# 2. holiday o or 1.
# 3. working day 0 or 1.
# 4. weathersit has three values good, bad and moderate weatherSit. 
# 5. season (1:spring, 2:summer, 3:fall, 4:winter)
# all the data seems to be in range
```

```python
Bike_df.season.replace({1:"spring", 2:"summer", 3:"fall", 4:"winter"},inplace = True)
Bike_df.weathersit.replace({1:'good',2:'moderate',3:'bad',4:'severe'},inplace = True)

Bike_df.mnth = Bike_df.mnth.replace({1: 'jan',2: 'feb',3: 'mar',4: 'apr',5: 'may',6: 'jun',
                  7: 'jul',8: 'aug',9: 'sept',10: 'oct',11: 'nov',12: 'dec'})

Bike_df.weekday = Bike_df.weekday.replace({0: 'sun',1: 'mon',2: 'tue',3: 'wed',4: 'thu',5: 'fri',6: 'sat'})
Bike_df.head()
```

# Dummy Variable Creation

```python
# Creating dummy variable for categorical columns
#["season","mnth","weekday","weathersit"]
#Creating  a new dataframe with the dummies
```

```python
Bike_df1 = pd.get_dummies(data=Bike_df,columns=["season","mnth","weekday","weathersit"],drop_first=True)

```

```python
Bike_df1.columns
```

```python
Bike_df1.shape
```

```python
Bike_df1.head()
```

# Train-Test Split

```python
# As the output column is cnt it is the y variable  and the rest of the columns will be the x variable
# train-test  70-30 

X=Bike_df1.drop(["cnt"],axis=1)
y=Bike_df1["cnt"]
```

```python
# importing the library for train_test_split
from sklearn.model_selection import train_test_split

# We specify this so that the train and test data set always have the same rows, respectively
np.random.seed(0)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=100)
```

```python
X.head()
```

```python
# Checking shape and size for train and test
print(X_train.shape)
print(X_test.shape)
```

# Scaling

```python
# Let us scale continuous variables
# Importing library for scaling 

from sklearn.preprocessing import MinMaxScaler
```

```python
# Applying min max scaler as it will removing the outliers if there are any as it will fit all the values between 0 and 1
scaler = MinMaxScaler()
```

```python
#Fit and transform training set only
X_train = scaler.fit_transform(X_train)
X_test=scaler.transform(X_test) # As test data is unseen
```

```python
# scaling method will return the array and we are converting those arrays in to data frame
X_train_df=pd.DataFrame(X_train,columns=X.columns)
X_test_df=pd.DataFrame(X_test,columns=X.columns)
```

```python
X_train_df.describe()
```

```python
X_train_df.head()
```

```python
# Let's check the correlation coefficients to see which variables are highly correlated
        
plt.figure(figsize = (30, 30))
sns.heatmap(X_train_df.corr(), annot = True, cmap="YlGnBu")
plt.show()
```

# Feature Selection

```python
# importing necessary libraries for feature selection

from sklearn.feature_selection import RFE
from sklearn.linear_model import LinearRegression
```

```python
# 15 features will be selected by auto feature selection

estimator=LinearRegression()
selector=RFE(estimator,n_features_to_select=16) #8-20
```

```python
selector=selector.fit(X_train_df,y_train)
selector.support_
```

```python
# columns selected automatically for modelling

selected_features=X_train_df.columns[selector.support_]
print(selected_features)
```

```python
X_train=X_train_df[selected_features]
X_test=X_test_df[selected_features]
```

```python
print(X_train.shape)
print(X_test.shape)
```

# Modelling

```python
# importing  libraries for modelling as stats model gives p value
import statsmodels.api as sm
```

```python
# Building Model 
# adding constant 

X_train_sm=sm.add_constant(X_train)
X_test_sm=sm.add_constant(X_test)
print(X_train_sm.shape)
print(X_test_sm.shape)
```

```python
# Ordinary Least Squares

model1=sm.OLS(np.array(y_train),X_train_sm)
res1=model1.fit()
res1.summary()
```

```python
# Acceptable Range for p-values is <0.05
# Acceptable Range for VIF is <=5
```

```python
# VIF 

from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data=pd.DataFrame()
vif_data["Feature"]=X_train_sm.columns

vif_data["VIF"]=[variance_inflation_factor(X_train_sm.values,i) 
                 for i in range(len(X_train_sm.columns))]
vif_data
```

# Manual feature Selection

```python
# Dropping the highest p score column which is not in range

X_train_sm=X_train_sm.drop("mnth_nov",axis=1)
X_test_sm=X_test_sm.drop("mnth_nov",axis=1)
```

```python
# model 2

model2=sm.OLS(np.array(y_train),X_train_sm)
res2=model2.fit()
res2.summary()
```

```python
# VIF 

from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data=pd.DataFrame()
vif_data["Feature"]=X_train_sm.columns

vif_data["VIF"]=[variance_inflation_factor(X_train_sm.values,i) 
                 for i in range(len(X_train_sm.columns))]
vif_data
```

```python


# Dropping the highest VIF column which is not in range

X_train_sm=X_train_sm.drop("weathersit_good",axis=1)
X_test_sm=X_test_sm.drop("weathersit_good",axis=1)
```

```python
# model 3

model3=sm.OLS(np.array(y_train),X_train_sm)
res3=model3.fit()
res3.summary()
```

```python
# VIF 

from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data=pd.DataFrame()
vif_data["Feature"]=X_train_sm.columns

vif_data["VIF"]=[variance_inflation_factor(X_train_sm.values,i) 
                 for i in range(len(X_train_sm.columns))]
vif_data
```

```python
# Dropping the highest p score column which is not in range

X_train_sm=X_train_sm.drop("season_spring",axis=1)
X_test_sm=X_test_sm.drop("season_spring",axis=1)
```

```python
# model 4

model4=sm.OLS(np.array(y_train),X_train_sm)
res4=model4.fit()
res4.summary()
```

```python
# VIF 

from statsmodels.stats.outliers_influence import variance_inflation_factor

vif_data=pd.DataFrame()
vif_data["Feature"]=X_train_sm.columns

vif_data["VIF"]=[variance_inflation_factor(X_train_sm.values,i) 
                 for i in range(len(X_train_sm.columns))]
vif_data
```

```python
print(X_train_sm.shape)
print(X_test_sm.shape)
print(y_train.shape)
```

```python
#From the above  Model Summary Report we can say that all the p-values of respective features are less than 0.05.
```

# Residual Analysis


```python
# Lets perform Residual Analysis

y_train_pred=res4.predict(X_train_sm)
residual =  (y_train- y_train_pred)
```

```python
#1. Assumption of Normally Distributed Error Term

# Plot the histogram of the error terms
sns.distplot(residual)
plt.xlabel('Residual (Error Term)', fontdict={'fontsize': 20})
plt.title('Error Distribution', fontdict={'fontsize': 25})
plt.show() 
```

```python
#Inferences: From the above graph it is evident that Error Distribution Is Normallly Distributed Across 0, which indicates that our model has handled the assumption of Error Normal Distribution properly.
```

```python
# Actual vs Predicted
c = [i for i in range(0,len(X_train_sm),1)]
plt.plot(c,y_train, color="blue")
plt.plot(c,y_train_pred, color="red")
plt.suptitle('Actual vs Predicted', fontsize = 15)
plt.xlabel('Index')
plt.ylabel('Demands')
plt.show()
```

#Actual and Predicted result following almost the same pattern so this model seems ok


```python
from sklearn.metrics import r2_score
y_test_pred=res4.predict(X_test_sm)
r2_train = r2_score(y_train, y_train_pred)
r2_test = r2_score(y_test, y_test_pred)
print("Train:",round(r2_train,3),"Test:",round(r2_test,3))
```

#R2 score is almost same for the test and train model

```python
fig = plt.figure()
plt.scatter(y_test_pred, y_test)
fig.suptitle('y_test vs y_pred', fontsize = 15)             
plt.xlabel('y_test', fontsize = 14)                          
plt.ylabel('y_pred', fontsize = 12) 
```

```python
#`Inference`_
# We can observe that variance of the residuals (error terms) is constant across predictions.  i.e error term does not vary much as the value of the predictor variable changes.
```

```python
# n is number of rows in X

n = X_test_sm.shape[0]


# Number of features (predictors, p) is the shape along axis 1
p = X_test_sm.shape[1]

# We find the Adjusted R-squared using the formula

test_adjusted_r2 = 1-(1-r2_train)*(n-1)/(n-p-1)
test_adjusted_r2
```

```python
#conclusion

# Train R-squire=0.82
# Test R-squire=0.803
# Train Adj-R-squire=0.815
# Test Adj-R-squire=0.807
#  seems a very good model.
```
